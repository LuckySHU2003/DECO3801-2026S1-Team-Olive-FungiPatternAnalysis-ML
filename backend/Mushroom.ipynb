# Data-record-mfs2-2026-03-13-04-32-17 Lions mane c2000-Studio3

```python
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# load the data
df1 = pd.read_excel("record-mfs2-2026-03-13-04-32-17 Lions mane c2000-Studio3.xlsx", sheet_name="record-mfs2-2026-03-13-04-32-17")

# view first few rows of the data
df1.head(5)
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
plt.plot(df1["ADC1 (green)"])
plt.plot(df1["ADC2 (yellow)"])
plt.plot(df1["ADC3 (orange)"])
plt.plot(df1["ADC4 (red)"])

plt.title("ADC Signal")
plt.show()
```

```python
# Regression Plot
y = df1["ADC1 (green)"].values
x = np.arange(len(y))

# Perform linear regression (fit a straight line)
coef = np.polyfit(x, y, 1)
y_pred = np.polyval(coef, x)

plt.figure(figsize=(10,4))
plt.scatter(x, y, s=5, label="ADC1 Signal")
plt.plot(x, y_pred, linewidth=2, label="Regression Line")

plt.title("ADC1 Regression Plot")
plt.xlabel("Time")
plt.ylabel("ADC1")
plt.legend()
plt.show()
```
